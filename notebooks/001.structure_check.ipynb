```python
print("=" * 60)
print("Step 3: Paragraph Analysis - Text -> CSV Mapping")
print("=" * 60)

# Load mapping and CSV metadata
mapping_path = "/tmp/mappings_volume/test_mapping.json"
with open(mapping_path, 'r') as f:
    mapping = json.load(f)

first_mapping = mapping['mappings'][0]
paragraph = first_mapping['paragraphs'][0]

print(f"Paragraph:")
print(f"{paragraph[:300]}..")

# Load CSV to get metadata
csv_filename = first_mapping['csv_files'][0]
csv_path = f"/Volumes/my_catalog/agentic_quality_check_dev/csvs_volume/{csv_filename}"
df = pd.read_csv(csv_path)

print(f"\n\n" + "=" * 60)

# Define phrases patters that hint at metrics

phrase_to_metric = {
    "average score": "mtc_score_average", 
    "took the check": "completed_check_pupil_percent",
    "working below": "working_below_pupil_percent", 
    "achieved this score": "mtc_score_25_pupil_percent"
}

# Find metrics are mentioned in paragraph
metrics_found = []
for phrase, column in phrase_to_metric.items():
    if phrase.lower() in paragraph.lower():
        # Check if column actually exists in CSV 
        if column in df.columns:
            metrics_found.append(column)
            print(f"Found: '{phrase}' -> {column}")
        else:
            print(f"Found phrase '{phrase} byt column {column} NOT in CSV")

print(f"\n\n" + "=" * 60)
print("STEP 3B: Identify Filters from Text")
print("=" * 60)

# Extract filter hints from paragraph

filters = {}

# Sex filter (boys/girls mentioned)
if 'boys' in paragraph.lower() and 'girls' in paragraph.lower():
    filters['sex'] = ['Boys', 'Girls']
    print(" Filter: sex= ['Boys', 'Girls'] (from 'boys and 'girls' in text)")

# Geographic level (assume National if not specified)
if 'geographic_level' in df.columns:
    filters['geographic_level'] = ['National']
    print("Filter: geographical_level = ['National'](default)")

# Time period (include all years for trends)
if 'time_period' in df.columns:
    filters['time_period'] = df['time_period'].unique().tolist()
    print(f"✅ Filter: time_period={filters['time_period']} (all years for trends)")

# Filter to aggregate rows only (all other dimensions = 'Total')
# These columns need to be 'Total' for national aggregates
if 'disadvantage_status' in df.columns:
    filters['disadvantage_status'] = ['Total']
if 'ethnicity_major' in df.columns:
    filters['ethnicity_major'] = ['Total']
if 'ethnicity_minor' in df.columns:
    filters['ethnicity_minor'] = ['Total']
if 'first_language' in df.columns:
    filters['first_language'] = ['Total']
if 'fsm_status' in df.columns:
    filters['fsm_status'] = ['Total']
if 'month_of_birth' in df.columns:
    filters['month_of_birth'] = ['Total']
if 'sen_provision' in df.columns:
    filters['sen_provision'] = ['Total']
if 'sen_primary_need' in df.columns:
    filters['sen_primary_need'] = ['Total']
if 'establishment_type_group' in df.columns:
    filters['establishment_type_group'] = ['All state funded']
print(f"✅ Filter: Aggregate rows only (all breakdowns = 'Total')")

print(f"\n\n" + "=" * 60)
print("STEP 3C: Apply Filters to CSV")
print("=" * 60)

# Start with full dataframe
df_filtered = df.copy()

# Apply each filter
for col, values in filters.items():
    if col in df_filtered.columns:
        df_filtered = df_filtered[df_filtered[col].isin(values)]
        print(f"✅ Filtered {col}: {len(df_filtered)} rows remain")

# Select only needed columns
columns_to_keep = ['sex', 'time_period'] + metrics_found
df_filtered = df_filtered[columns_to_keep]

print(f"\n📋 Final columns: {df_filtered.columns.tolist()}")
print(f"📊 Filtered data: {df_filtered.shape[0]} rows × {df_filtered.shape[1]} columns")

print(f"\n\n" + "=" * 60)
print("FILTERED DATA (Long Format)")
print("=" * 60)
display(df_filtered.head(10))

print("\n✅ STEP 3 Complete!")
print("\n🎯 Next: Apply melt+pivot to convert to wide format")

```

```python
print("=" * 60)
print("STEP 4: Melt + Pivot Transformation")
print("=" * 60)

# Step 4A: Melt the metric columns into long format
print("\nSTEP 4A: Melt metric columns...")

# Identify which columns are metrics vs identifiers
id_cols = ['time_period', 'sex']
metric_cols = metrics_found  # From previous cell

# Melt: Convert metric columns → rows
df_melted = df_filtered.melt(
    id_vars=id_cols,
    value_vars=metric_cols,
    var_name='metric',
    value_name='value'
)

print(f"✅ Melted: {df_melted.shape[0]} rows × {df_melted.shape[1]} columns")
print("\nSample melted data:")
display(df_melted.head(8))

# Step 4B: Pivot to match your exact table structure
print("\n" + "=" * 60)
print("STEP 4B: Pivot to wide format (Architecture.md specification)...")
print("=" * 60)
print("Structure: rows=[metric, time_period], columns=sex")

# Pivot: Multi-level rows (metric, time_period), simple columns (sex)
df_wide = df_melted.pivot_table(
    index=['metric', 'time_period'],
    columns='sex',
    values='value',
    aggfunc='first'
)

print(f"\n✅ Pivoted: {df_wide.shape[0]} rows (metric×time) × {df_wide.shape[1]} columns (sex)")
print(f"\n📋 Row index: multi-level (metric, time_period)")
print(f"📋 Columns: {df_wide.columns.tolist()}")

# Rename metric labels for readability (without % in the name)
metric_labels = {
    'mtc_score_average': 'Average attainment score',
    'completed_check_pupil_percent': 'Percentage of eligible pupils who took the check',
    'working_below_pupil_percent': 'Percentage of pupils working below the national curriculum expectation'
}

df_wide_labeled = df_wide.rename(index=metric_labels, level=0)

print("\n" + "=" * 60)
print("FINAL WIDE-FORMAT TABLE")
print("=" * 60)
print("\nRaw metric names:")
# Reset index to make metrics visible as columns
df_wide_display = df_wide.reset_index()
# Show metric name only once per group
df_wide_display_formatted = df_wide_display.copy()
df_wide_display_formatted['metric'] = df_wide_display_formatted['metric'].mask(
    df_wide_display_formatted['metric'].duplicated(), ''
)
display(df_wide_display_formatted)

print("\n" + "=" * 60)
print("With readable labels (matches your PDF table):")
print("=" * 60)
df_wide_labeled_display = df_wide_labeled.reset_index()
# Show metric name only once per group
df_wide_labeled_display_formatted = df_wide_labeled_display.copy()

# Add % sign to percentage metrics
for idx, row in df_wide_labeled_display_formatted.iterrows():
    metric_name = row['metric']
    if '%' in str(metric_name) or 'percent' in str(metric_name).lower():
        # Add % to Boys and Girls columns
        if pd.notna(row['Boys']):
            df_wide_labeled_display_formatted.at[idx, 'Boys'] = str(row['Boys']) + '%'
        if pd.notna(row['Girls']):
            df_wide_labeled_display_formatted.at[idx, 'Girls'] = str(row['Girls']) + '%'

# Show metric name only once per group (after adding %)
df_wide_labeled_display_formatted['metric'] = df_wide_labeled_display_formatted['metric'].mask(
    df_wide_labeled_display_formatted['metric'].duplicated(), ''
)
display(df_wide_labeled_display_formatted)

print("\n✅ STEP 4 Complete!")
print("\n🎯 Table structure: Metrics as ROWS, (Time Period × Sex) as COLUMNS")
print("📖 This matches Architecture.md Phase 6 specification!")
```

```python
print("=" * 60)
print("PHASE 8: VERIFICATION LOGIC")
print("=" * 60)
print("\n🎯 Goal: Verify paragraph claims match extracted table data")
print("\n" + "=" * 60)

# Step 1: Define claims from the paragraph
print("\nSTEP 1: Extract Claims from Paragraph")
print("=" * 60)

paragraph = first_mapping['paragraphs'][0]
print(f"\nParagraph:\n{paragraph[:200]}...\n")

# Define expected claims (manually for now, later can be LLM-parsed)
claims = [
    {
        "description": "Girls took the check: 97%",
        "metric": "Percentage of eligible pupils who took the check",
        "time_period": 202122,  # Most recent year mentioned in context
        "gender": "Girls",
        "expected_value": "97%"
    },
    {
        "description": "Boys took the check: 95%",
        "metric": "Percentage of eligible pupils who took the check",
        "time_period": 202122,
        "gender": "Boys",
        "expected_value": "95%"
    },
    {
        "description": "Girls average score: 19.6",
        "metric": "Average attainment score",
        "time_period": 202122,
        "gender": "Girls",
        "expected_value": "19.6"
    },
    {
        "description": "Boys average score: 20.0",
        "metric": "Average attainment score",
        "time_period": 202122,
        "gender": "Boys",
        "expected_value": "20.0"
    }
]

print(f"\n✅ Defined {len(claims)} claims to verify:")
for i, claim in enumerate(claims, 1):
    print(f"  {i}. {claim['description']}")

# Step 2: Look up values in the extracted table
print("\n" + "=" * 60)
print("STEP 2: Look Up Values in Extracted Table")
print("=" * 60)

verification_results = []

for claim in claims:
    metric = claim['metric']
    time_period = claim['time_period']
    gender = claim['gender']
    expected = claim['expected_value']
    
    # Look up the value in df_wide_labeled_display_formatted
    try:
        # Find the row matching metric and time_period
        mask = (
            (df_wide_labeled_display['metric'] == metric) & 
            (df_wide_labeled_display['time_period'] == time_period)
        )
        matching_rows = df_wide_labeled_display[mask]
        
        if len(matching_rows) == 0:
            actual = "NOT FOUND"
            match = False
        else:
            # Get the value for the specific gender
            actual_raw = matching_rows.iloc[0][gender]
            
            # Format actual value to match expected format
            # If expected has %, add % to actual
            if '%' in expected:
                actual = str(actual_raw) + '%'
            else:
                actual = str(actual_raw)
            
            # Compare
            match = (actual == expected)
    
    except Exception as e:
        actual = f"ERROR: {e}"
        match = False
    
    verification_results.append({
        "Claim": claim['description'],
        "Expected": expected,
        "Actual": actual,
        "Match": "✅" if match else "❌"
    })

# Step 3: Display Verification Results
print("\n" + "=" * 60)
print("STEP 3: VERIFICATION RESULTS")
print("=" * 60)

df_verification = pd.DataFrame(verification_results)
display(df_verification)

# Summary
total_claims = len(verification_results)
matches = sum(1 for r in verification_results if r['Match'] == '✅')
mismatches = total_claims - matches

print("\n" + "=" * 60)
print("SUMMARY")
print("=" * 60)
print(f"\n📊 Total Claims: {total_claims}")
print(f"✅ Matches: {matches}")
print(f"❌ Mismatches: {mismatches}")

if mismatches == 0:
    print("\n🎉 SUCCESS! All claims verified!")
    print("✅ The paragraph text matches the extracted table data.")
else:
    print("\n⚠️  WARNING! Quality issues detected!")
    print(f"❌ {mismatches} claim(s) do not match the table data.")
    print("\n🔍 Review the mismatches above to identify errors.")

print("\n✅ PHASE 8 Complete!")
print("\n🎯 This is the core of your automated quality assurance system!")
```
